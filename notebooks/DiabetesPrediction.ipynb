```python
# Imports bsicos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, roc_curve, classification_report
)
import joblib

# Opcional: para mostrar grficos inline
%matplotlib inline
sns.set(style="whitegrid")
RANDOM_STATE = 42
```

```python
df = pd.read_csv("diabetes.csv")
print("Shape:", df.shape)
df.head()
```

```python
df.info()
df.describe().T
df['Outcome'].value_counts()
```

```python
# Conteo de clases
sns.countplot(x='Outcome', data=df)
plt.title('Distribucin Outcome')
plt.show()

# Histograma de glucosa
sns.histplot(df['Glucose'], kde=True)
plt.title('Distribucin Glucose')
plt.show()

# Heatmap de correlacin
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlacin entre variables")
plt.show()
```

```python
cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[cols_with_zero] = df[cols_with_zero].replace(0, np.nan)

# Mostrar cuantos NaN por columna
df[cols_with_zero].isna().sum()
```

```python
imputer = SimpleImputer(strategy='median')
df[cols_with_zero] = imputer.fit_transform(df[cols_with_zero])

# Verificar
df.isna().sum()
```

```python
X = df.drop('Outcome', axis=1)
y = df['Outcome']

# train/test stratificado (mantiene proporcin de clases)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)

print(X_train.shape, X_test.shape)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000, random_state=RANDOM_STATE),
    "RandomForest": RandomForestClassifier(n_estimators=100, random_state=RANDOM_STATE),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=RANDOM_STATE)
}

results = []
for name, model in models.items():
    # elegir datos escalados para regresin, pero rboles funcionarn con los mismos datos escalados (ok)
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    y_prob = model.predict_proba(X_test_scaled)[:,1]

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)   # sensibilidad
    f1 = f1_score(y_test, y_pred)
    roc = roc_auc_score(y_test, y_prob)

    # confusion matrix para especificidad
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    specificity = tn / (tn + fp)

    results.append([name, acc, prec, rec, specificity, f1, roc])

results_df = pd.DataFrame(results, columns=['Model','Accuracy','Precision','Sensitivity','Specificity','F1','ROC-AUC'])
results_df
```

```python
# Celda 9  Curva ROC comparativa
plt.figure(figsize=(8,6))
for name, model in models.items():
    y_prob = model.predict_proba(X_test_scaled)[:,1]
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=f"{name} (AUC={roc_auc_score(y_test,y_prob):.3f})")

plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves")
plt.legend()
plt.show()
```

```python
# Celda 10  Seleccionar mejor modelo y guardar
# suponiendo que queremos el modelo con mayor ROC-AUC
best_name = results_df.sort_values('ROC-AUC', ascending=False).iloc[0]['Model']
best_model = models[best_name]
print("Best:", best_name)

# Guardar modelo y scaler
joblib.dump(best_model, "model_best.pkl")
joblib.dump(scaler, "scaler.pkl")
```

```python
# Celda 11  Importancia de features (RandomForest / XGBoost)
# Usar RandomForest para explicar (si es el mejor)
rf = models['RandomForest']
importances = rf.feature_importances_
feat_imp = pd.Series(importances, index=X.columns).sort_values(ascending=False)
feat_imp.plot(kind='bar')
plt.title("Feature importances (RandomForest)")
plt.show()

# Alternativa: usar SHAP para interpretabilidad ms fina
# pip install shap
import shap
explainer = shap.TreeExplainer(rf)
shap_values = explainer.shap_values(X_test)
shap.summary_plot(shap_values, X_test, feature_names=X.columns)
```

```python
# Celda 12  Ajuste de umbral (threshold tuning)
# A veces quieres priorizar sensibilidad (detectar todos los enfermos). Puedes elegir un umbral distinto de 0.5:
#
from sklearn.metrics import precision_recall_curve

y_prob = best_model.predict_proba(X_test_scaled)[:,1]
precisions, recalls, thresholds = precision_recall_curve(y_test, y_prob)

# Ejemplo: encontrar umbral que maximice (recall + specificity) o max F1
f1_scores = 2 * (precisions * recalls) / (precisions + recalls + 1e-8)
best_idx = np.argmax(f1_scores)
best_thresh = thresholds[best_idx]
print("Best threshold (F1):", best_thresh)

# Aplicar y evaluar
y_pred_thresh = (y_prob >= best_thresh).astype(int)
print(classification_report(y_test, y_pred_thresh))
```
